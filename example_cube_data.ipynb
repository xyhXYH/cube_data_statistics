```python
import pandas as pd
import numpy as np
```

```python
data_generator = lambda: np.array([np.random.rand(12), np.random.randint(1, 10, 12), np.random.randint(20, 30, 12)]).transpose()
```

```python
sample = np.array([data_generator() for _ in range(4)])
```

```python
sample.shape
```

## 4个样本， 12个月， 3个feature

```python
sample
```

# 层次索引之后，每个listing的每个月的第一个feature的值

```python
pd.DataFrame(sample[:,:,0]).stack()
```

# 以此类推

```python
columns = ['feature1', 'feature2', 'feature3']
df = pd.DataFrame({columns[i]: pd.DataFrame(sample[:,:,i]).stack() for i in range(len(columns))})
```

```python
df
```

# 给index标上名字，便于等一下groupby

```python
indexes = df.index
indexes
```

```python
# 以这样的方式可以命名
indexes.set_names(['listing', 'month'])
```

```python
# 所以：
df.index = df.index.set_names(['listing', 'month'])
```

```python
df
```

# 万事俱备，聚合吧

```python
# 参数level必须要， groupby的第一个变量是 by，是指列名
df.groupby(level='listing').count()
#每个listing的每个变量有12个月的维度， 所以count为12
```

```python
df.groupby(level='month').count()
#同理，4个listing
```

# 其他 sum、 average等都类似，参照groupby之后的操作

## 下面是按条件聚合

```python
# 每个listing，第二个变量，大于8的月数
result = df[df['feature2'] > 8].groupby(level='listing').apply(len)
print(type(result))
result

# 结果对不对，自己去上面数
```

# 那么，如何把没有的这个条件的listing补齐？
> 原因：`df[df['feature2'] > 8]` 这里就已经把不符合条件的行删掉了，因此listing==1的行都不满足，都被删掉

# 如果期待的应该是listing长度不变，缺失值补na？

```python
listing_index = df.index.get_level_values('listing').unique()
listing_index
```

```python
result_pad = pd.Series(result, index=listing_index)
result_pad
```
